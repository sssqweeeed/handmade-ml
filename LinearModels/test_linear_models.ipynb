```python
import LinearReg
import seaborn as sns
import pandas as pd
```

```python
def generate_data(k = -0.5, b = 10, a_range = -1.0, b_range = 1.0, n = 10, noise_scale=0.3):
    x_values = np.linspace(a_range, b_range, num=n)
    y_values = k*x_values + b# + np.random.normal(size=x_values.size)
    if noise_scale > 0:
        noise = np.random.normal(scale=noise_scale, size=n)
        y_values += noise
    data = np.column_stack((x_values, y_values)).T
    return data
data = generate_data()
sns.scatterplot(x = data[0], y = data[1])
```

```python
model = LinearReg.Ridge()
model.fit(data[0, np.newaxis].T, data[1])
pred_y = model.predict(data[0, np.newaxis].T)

sns.scatterplot(x = data[0], y = pred_y.reshape(-1, ))
sns.scatterplot(x = data[0], y = data[1])
```
